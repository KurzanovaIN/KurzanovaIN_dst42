## Решение задачи:
Пробуем уменьшить среднее число попыток для угадывания.
Используется 2 варианта решения.

```python
#Вариант1.
import numpy as np

def game_core_v2(number):
    '''Сначала устанавливаем любое random число, а потом уменьшаем 
    или увеличиваем его в зависимости от того, больше оно 
    или меньше нужного.
       Функция принимает загаданное число и возвращает число попыток
    Находим верхнюю границу предполагаемого диапазона.
    Проверяем загаданное число, если оно больше предполагаемого, 
    увеличиваем предполагаемое число. После этого проверяем не стало ли
    предполагаемое число больше загаданного.
       Если при проверке число меньше предполагаемого, уменьшаем верхнюю
    границу предполагаемого и увеличиваем счетчик попыток.
    Иначе: если число меньше предполагаемого, уменьшаем предполагаемый диапазон.
       Если при проверке число становится больше предполагаемого, увеличиваем
    нижнюю границу диапазона и увеличиваем счетчик попыток'''
    count = 1                           # счетчик попыток
    predict = np.random.randint(1,101)  # предполагаемое число
    while number != predict:   # number - число, загаданное компьютером
        count+=1
        if number > predict: 
            predict = (predict+100) // 2   
            if number < predict:
                while number != predict:
                    count+=1
                    predict-=1
                break
        else:
            predict = predict//2
            if number > predict:
                while number != predict:
                    count+=1
                    predict+=1
                break
    #print(number,predict) #выведено на экран для проверки 
    return(count) # выход из цикла, если угадали


def score_game(game_core):
    '''Запускаем игру 1000 раз, чтобы узнать, как быстро игра угадывает число'''
    count_ls = []
    np.random.seed(1)  # фиксируем RANDOM SEED, чтобы ваш эксперимент был воспроизводим!
    random_array = np.random.randint(1, 101, size=(233))

    for number in random_array:
        count_ls.append(game_core(number))
    score = int(np.mean(count_ls))
    print(f"Ваш алгоритм угадывает число в среднем за {score} попыток")
    return(score)

# Проверяем
score_game(game_core_v2)
```

**Итог:** при реализации кода из Варианта1, программа работает только при числе
запусков игры size=(233), при превышении данного числа запусков, программа перестает считать. Алгоритм угадывает число в среднем за 12 попыток.

Продолжим уменьшать число угадываний, учитывая число запусков 1000

```python
#Вариант2.
import numpy as np


def game_core_v2(number):
    '''Сначала устанавливаем любое random число, а потом уменьшаем 
    или увеличиваем его в зависимости от того, больше оно 
    или меньше нужного.
       Функция принимает загаданное число и возвращает число попыток.
       Алгоритм условия цикла: если загаданное число больше предполагаемого
    присваиваем нижней границе диапазона значений число, следующее за 
    предполагаемым числом, иначе присваиваем верхней границе диапазона число, 
    равное предполагаемому.
       Затем расчитываем новое число, равное усредненному между начальным и
    конечным значениями найденного диапазона.   
       Повторять до тех пор, пока число не будет угадано.'''
    count = 1                           # счетчик попыток
    predict = np.random.randint(1,101)  # предполагаемое число
    start_interval = 1    #диапазон предполагаемых чисел: начало диапазона
    end_interval = 100    #диапазон предполагаемых чисел: конец диапазона
    while number != predict:   # number - число, загаданное компьютером
        count+=1
        if number > predict: 
            start_interval = predict+1
            #print("start_interval",start_interval)
        else:
            end_interval = predict
            #print("end_interval",end_interval)
        predict = (start_interval+end_interval)//2    
        #print(number,predict) # вывод на экран для проверки
    return(count) # выход из цикла, если угадали


def score_game(game_core):
    '''Запускаем игру 1000 раз, чтобы узнать, как быстро игра угадывает число'''
    count_ls = []
    np.random.seed(1)  # фиксируем RANDOM SEED, чтобы ваш эксперимент был воспроизводим!
    random_array = np.random.randint(1, 101, size=(1000))

    for number in random_array:
        count_ls.append(game_core(number))
    score = int(np.mean(count_ls))
    print(f"Ваш алгоритм угадывает число в среднем за {score} попыток")
    return(score)

# Проверяем
score_game(game_core_v2)
```

**Итог:** по данному алгоритму программа находит верное число в среднем за 6 попыток. Условие запуска игры 1000 раз соблюдается, программа работает также и при количестве запусков size=(10000), средний результат при этом не меняется (в среднем 6 попыток).
